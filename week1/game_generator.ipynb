Voy a crear una aplicación basada en IA generativa (Claude) que sirva para generar pequeños juegos con Tecnología Web

```python
import os
from dotenv import load_dotenv
import anthropic
from openai import OpenAI
from IPython.display import Markdown, display, update_display
import pygame
```

```python
load_dotenv(override=True)
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
openai_api_key = os.getenv('OPENAI_API_KEY')

if anthropic_api_key:
    print(f"Anthropic API Key exists and begins {anthropic_api_key[:7]}")
else:
    print("Anthropic API Key not set")

if openai_api_key and openai_api_key.startswith('sk-proj-') and len(openai_api_key)>10:
    print("OpenAI API key looks good so far")
else:
    print("There might be a problem with your OpenAI API key? Please visit the troubleshooting notebook!")
```

```python
ANTRHOPIC_MODEL = 'claude-3-7-sonnet-latest'
OPENAI_MODEL = 'gpt-4o-mini'

claude = anthropic.Anthropic()
openai = OpenAI()
```

```python
system_message = "You are a python developer specialized in creating video games. \
You will receive some specifications and build a game coded in python. \
You must response only the generated python code. Do not include any markdown, code block indicators or description. Try not to generate more than 10000 tokens"

user_prompt = "I want a RPG game with a fantastic ambientation. Easy to play and with vibrant colors"
```

```python
# Claude 3.7 Sonnet
# API needs system message provided separately from user prompt
# Also adding max_tokens
def anthropic_generation():
    result = claude.messages.stream(
        model=ANTRHOPIC_MODEL,
        max_tokens=64000,
        system=system_message,
        messages=[
            {"role": "user", "content": user_prompt}
        ],
    )
    
    reply = ""
    with result as stream:
        for text in stream.text_stream:
            reply += text
            print(text, end="", flush=True)
    return reply
```

```python
# GPT o4 mini

messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
]
def openai_generation():
    stream = openai.chat.completions.create(
        model=OPENAI_MODEL,
        messages=messages,
        stream=True
    )

    reply = ""
    for chunk in stream:
        fragment = chunk.choices[0].delta.content or ""
        reply += fragment
        print(fragment, end='', flush=True)
    exec(reply)
```

```python
openai_generation()
```
